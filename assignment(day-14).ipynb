```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv(r'C:\Users\nutha\Downloads\Day_14_Pharma_data.csv')
data
```

```python
# Check for missing values
print(data.isnull().sum())
```

```python
# Drop rows with missing values or handle them accordingly
data.dropna(inplace=True)
data
```

```python
# Check for duplicates
print(data.duplicated().sum())
```

```python
# Drop duplicates if any
data.drop_duplicates(inplace=True)
data
```

```python
# Check column names
print(data.columns)
```

```python
# Check for missing values in the 'Product' column
print(data['Product_ID'].isnull().sum())
```

```python
# Violin plot: Distribution of Effectiveness and Side Effects for each Product
plt.figure(figsize=(12, 6))
sns.violinplot(data=data, x='Product_ID', y='Effectiveness', hue='Side_Effects', split=True, palette='coolwarm')
plt.title('Distribution of Effectiveness and Side Effects for Each Product')
plt.xlabel('Product_ID')
plt.ylabel('Effectiveness')
plt.show()
```

```python
# Pairplot: Explore relationships between Effectiveness, Side_Effects, and Marketing_Spend
sns.pairplot(data[['Effectiveness', 'Side_Effects', 'Marketing_Spend']])
plt.title('Pairplot: Effectiveness, Side Effects, and Marketing Spend')
plt.show()
```

```python
# Boxplot: Effectiveness for different trial periods
plt.figure(figsize=(12, 6))
sns.boxplot(data=data, x='Trial_Period', y='Effectiveness')
plt.title('Effectiveness Across Different Trial Periods')
plt.xlabel('Trial Period')
plt.ylabel('Effectiveness')
plt.show()
```

```python
# Regression plot: How Marketing Spend affects drug Effectiveness
plt.figure(figsize=(12, 6))
sns.regplot(data=data, x='Marketing_Spend', y='Effectiveness', scatter_kws={'s': 10}, line_kws={'color': 'red'})
plt.title('Marketing Spend vs Drug Effectiveness')
plt.xlabel('Marketing Spend')
plt.ylabel('Effectiveness')
plt.show()
```
